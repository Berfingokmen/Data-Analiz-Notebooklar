```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(1,6)
```

```python
x
```

```python
y = np.arange(2,11,2)
```

```python
y
```

```python
plt.plot(x,y,"red")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("X vs Y Graph")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"blue")
plt.subplot(2,2,2)
plt.plot(y,x,"yellow")
plt.subplot(2,2,3)
plt.plot(y,x,"red")
plt.subplot(2,2,4)
plt.plot(x,x ** 2,"black")
plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.2,0.3])

axes2.plot(x,y)
axes2.set_xlabel("Inner X")
axes2.set_ylabel("Inner Y")
axes2.set_title("Inner Graph")

axes1.plot(y,x)
axes1.set_xlabel("Outer X")
axes1.set_ylabel("Outer Y")
axes1.set_title("Outer Graph")
```

```python
x
```

```python
y
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1)

axes[0].plot(x,y)
axes[0].set_title("First Axes")
axes[1].plot(x,x ** 3)
axes[1].set_title("Second Axes")

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1,figsize =(8,6))

axes[0].plot(x,y,color = "red")
axes[0].set_title("First Axes")
axes[1].plot(x,x ** 0.5,color ="purple")
axes[1].set_title("Second Axes")

plt.tight_layout()
```

```python
fig = plt.figure(figsize =(6,4))

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x ** 0.5,color = "red",label = "X Karekök")
axes.plot(x,x ** 2,color = "yellow",label = "X Kare")
axes.plot(x, x ** 3,color = "#4055c2",label = "X Küp")

axes.legend()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,color = "red",lw = 3,marker = "o",markersize = 20,markerfacecolor = "black"
         ,markeredgecolor = "yellow",markeredgewidth =3 )
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,color = "red",lw = 3,marker = "o",markersize = 10,markerfacecolor = "black"
         ,markeredgecolor = "yellow",markeredgewidth =3 )

axes.set_xlim(0,10)
axes.set_ylim(0,50)
```
